```python
import boto3

# Criando uma sessão com o perfil 'cli-user'
session = boto3.Session(profile_name='cli-user', region_name='us-east-1')

# Criando o cliente S3 com a sessão configurada
s3 = session.client('s3')

# Listando os arquivos no bucket
bucket_name = 'vendas-etl-projeto'
response = s3.list_objects_v2(Bucket=bucket_name)

# Exibindo o conteúdo do bucket
for obj in response.get('Contents', []):
    print(obj['Key'])
```

```python
import pandas as pd
from io import StringIO

# Função para baixar o arquivo CSV do S3 e carregá-lo em um DataFrame
def download_csv_from_s3(bucket, file_name):
    # Baixando o objeto do S3
    response = s3.get_object(Bucket=bucket, Key=file_name)
    
    # Lendo o conteúdo do arquivo CSV e carregando no Pandas DataFrame
    content = response['Body'].read().decode('utf-8')
    return pd.read_csv(StringIO(content))

# Baixando o dataset
df = download_csv_from_s3(bucket_name, 'vendas_dataset.csv')

# Mostrar as primeiras linhas do DataFrame
df.head()
```

```python
# Verificar as informações gerais sobre o DataFrame
df.info()

# Exibir as primeiras linhas
df.head()
```

```python
# Convertendo a coluna 'OrderDate' para datetime
df['OrderDate'] = pd.to_datetime(df['OrderDate'], errors='coerce')

# Verificando as primeiras linhas após a conversão
df.head()
```

```python
from sqlalchemy import create_engine

# Criar uma conexão com o banco de dados (MySQL)
# Substitua 'senha' pela sua senha real do MySQL
engine = create_engine('mysql+pymysql://root:3101@localhost:3306/vendas_dataset')

# Armazenar os dados no MySQL
df.to_sql('vendas_etl', con=engine, if_exists='replace', index=False)
```
